````python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import minimize
import matplotlib.pyplot as plt

from matplotlib import rc
import seaborn as sns

plt.style.use(['seaborn', "seaborn-dark"][0])
plt.rc('axes', titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rcParams.update({'font.size': 16})
plt.rcParams['axes.titlesize'] = 16
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams.update({'lines.markeredgewidth': 1})
plt.rcParams.update({'errorbar.capsize': 2})
````

## P1

### Ejercicio 1

````python
def dinamica(t, z):
    x, p = z[0], z[1]
    u = 0 if p+1>=0 else 1
    dx = x*u
    dp = 1-u*(p+1)
    return np.array([dx, dp])

def solve_pontryagin(x0, T):
    def error(p0, x0, T):
        z0 = np.array([x0, p0[0]])
        sol_z = solve_ivp(dinamica, t_span=[0, T], y0=z0)
        pT = sol_z.y[1, -1]
        return 0.5*pT*pT
    
    p0 = 1
    opt_p0 = minimize(error, x0=p0, args=(x0, T)).x
    z0 = np.array([x0, opt_p0[0]])
    sol_z = solve_ivp(dinamica, t_span=[0, T], y0=z0, dense_output=True)
    z = sol_z.sol
    x = lambda t: z(t)[0]
    p = lambda t: z(t)[1]
    u = lambda t: 0 if p(t)+1>=0 else 1
    return (x, p, u)
````

````python
def plot_p1(X0, T):
    fig, ax = plt.subplots(1, 3, figsize=(5*3, 5))
    solutions = {}
    for i in range(len(X0)):
        x0 = X0[i]
        x, p, u = solve_pontryagin(x0=x0, T=T)
        t_span = np.linspace(0, T)

        xt = x(t_span)
        pt = p(t_span)
        ut = [u(t) for t in t_span]
        
        ax[0].plot(t_span, xt, label=f"x0: {x0}")
        ax[1].plot(t_span, pt)
        ax[2].plot(t_span, ut);
        
        solutions[x0] = (t_span, xt, pt, ut)
    
    ax[0].set_xlabel("tiempo (t)")
    ax[1].set_xlabel("tiempo (t)")
    ax[2].set_xlabel("tiempo (t)");

    ax[0].set_title("estado (x)")
    ax[1].set_title("adjunto (p)")
    ax[2].set_title("control (u)");
    
    ax[0].legend()
    
    return solutions
````

````python
p1_T_5 = plot_p1(X0 = [0.5, 1, 5], T = 5)
plt.savefig("e1_1.pdf", bbox_inches='tight')
````

## P2

### Ejercicio 2

````python
def dt_V(s, y, T):
    a =  np.multiply(-y, np.exp(-s+T-1))
    b = -y
    return a*(s<T-1) + b*(s>=T-1)

def dx_V(s, y, T):
    a = np.exp(-s+T-1)
    b = (T-s)
    return a*(s<T-1) + b*(s>=T-1)

def check(s, y, T):
    return dt_V(s, y, T) + y + y * (dx_V(s, y, T)-1) * (dx_V(s, y, T)-1 > 0)
````

````python
import seaborn as sns
````

Valor de la ecuación

````python
T = 10
M = 1000
S = np.linspace(0, T, 100)
Y = np.linspace(0, M, 100)

SS = []
YY = []
for s in S:
    for y in Y:
        SS.append(s)
        YY.append(y)
        
val = check(s=np.array(SS), y=np.array(YY), T=T)
Z = np.ones((Y.shape[0], S.shape[0]))
for i in range(S.shape[0]):
    for j in range(Y.shape[0]):
        Z[j, i] = val[i*S.shape[0]+j]

sns.heatmap(Z, annot=False, cmap="vlag");
plt.xticks(range(len(S)), [int(S[i]) if i%10 in [0, len(S)%10] else "" for i in range(len(S))], rotation=0);
plt.xlabel("Tiempo (t)")

plt.yticks(range(len(Y)), [int(Y[i]) if i%10 in [0, len(Y)%10] else "" for i in range(len(Y))], rotation=0);
plt.ylabel("Espacio (y)");
plt.title("Valor");

plt.savefig("e2_1.pdf", bbox_inches='tight')
````

## Ejercicio 3

````python
def solve_hjb(x0, T):
    def dinamica_B(t, x):
        global T
        u = 0 if -dx_V(t, x, T)+1>0 else 1
        dx = x * u
        return np.array([dx])

    sol_x = solve_ivp(dinamica_B, t_span=[0, T], y0=np.array([x0]), dense_output=True)
    z = sol_x.sol
    x = lambda t: z(t)[0]
    u = lambda t: 0 if -dx_V(t, x(t), T)+1 > 0 else 1
    return (x, u)

def plot_p3(X0, T):
    fig, ax = plt.subplots(1, 2, figsize=(5*2, 5))
    solutions = {}
    for i in range(len(X0)):
        x0 = X0[i]
        x, u = solve_hjb(x0=x0, T=T)
        t_span = np.linspace(0, T)

        xt = x(t_span)
        ut = [u(t) for t in t_span]
        
        solutions[x0] = (t_span, xt, ut)
        
        ax[0].plot(t_span, xt, label=f"x0: {x0}")
        ax[1].plot(t_span, ut)

    ax[0].set_xlabel("tiempo (t)")
    ax[1].set_xlabel("tiempo (t)")

    ax[0].set_title("estado (x)")
    ax[1].set_title("control (u)");
    
    ax[0].legend()
    
    return solutions
````

````python
p3_T_5 = plot_p3(X0 = [0.5, 1, 5], T = 5)
plt.savefig("e3_1.pdf", bbox_inches='tight')
````

````python
p3_T_10 = plot_p3(X0 = [0.5, 1, 5], T = 10)
plt.savefig("e3_2.pdf", bbox_inches='tight')
````

````python
def plot_p4(sols):    
    fig, ax = plt.subplots(1, 3, figsize=(5*3, 5))
    solutions = {}
    for i in range(len(sols)):
        sol = sols[i]
        controls = sol.controls
        times = sol.discretization_times
        states = sol.states
        x0 = [0.5, 1, 5][i]
        ax[0].plot(times, states["x"].values, label=f"x0: {x0}")
        ax[1].plot(times, states["z"].values)
        ax[2].plot(times, controls["u"].values)
        solutions[x0] = (times, states["x"].values, states["z"].values, controls["u"].values)

    ax[0].set_xlabel("tiempo (t)")
    ax[1].set_xlabel("tiempo (t)")
    ax[2].set_xlabel("tiempo (t)")
    
    ax[0].set_title("estado (x)")
    ax[1].set_title("consumo (c)")
    ax[2].set_title("control (u)");
    
    ax[0].legend()
    return solutions
````

## Ejercicio 5

````python
def get_consumo(time_p, x_p, u_p):
    dt = time_p[1:] - time_p[:-1] 
    y = (1-u_p) * x_p
    consumo = np.sum( y[:-1] * dt )
    return consumo
````

````python
def plot_all(p1_T, p3_T, p4_T):
    fig, ax = plt.subplots(1, 2, figsize=(5*2, 5))
    consumos = []
    for i in range(3):
        x0 = [0.5, 1, 5][i]

        time_p1, x_p1, u_p1 = np.array(p1_T[x0][0]), np.array(p1_T[x0][1]), np.array(p1_T[x0][-1])
        time_p3, x_p3, u_p3 = np.array(p3_T[x0][0]), np.array(p3_T[x0][1]), np.array(p3_T[x0][-1])
        time_p4, x_p4, u_p4 = np.array(p4_T[x0][0]), np.array(p4_T[x0][1]), np.array(p4_T[x0][-1])

        c = ["tab:blue", "tab:green", "tab:red"][i]

        ax[0].plot(time_p1, x_p1, linestyle="dotted", c=c, label=f"x0: {x0}, PMP")
        ax[0].plot(time_p3, x_p3, linestyle="dashed", c=c, label=f"x0: {x0}, HJB")
        ax[0].plot(time_p4, x_p4, linestyle="dashdot", c=c, label=f"x0: {x0}, BOCOP")

        ax[1].plot(time_p1, u_p1, linestyle="dotted", c=c)
        ax[1].plot(time_p3, u_p3, linestyle="dashed", c=c)
        ax[1].plot(time_p4, u_p4, linestyle="dashdot", c=c)
        
        consumo_p1 = get_consumo(time_p1, x_p1, u_p1)
        consumo_p3 = get_consumo(time_p3, x_p3, u_p3)
        consumo_p4 = get_consumo(time_p4, x_p4, u_p4)
        consumos.append([consumo_p1, consumo_p3, consumo_p4])   

    ax[0].set_xlabel("tiempo (t)")
    ax[1].set_xlabel("tiempo (t)")

    ax[0].set_title("estado (x)")
    ax[1].set_title("control (u)");

    ax[0].legend()
    return consumos
````

````python
cc_T_5 = plot_all(p1_T_5, p3_T_5, p4_T_5)
plt.savefig("e5_1.pdf", bbox_inches='tight')
````

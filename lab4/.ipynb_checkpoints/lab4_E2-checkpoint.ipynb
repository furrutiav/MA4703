```python
import numpy as np
import scipy
from matplotlib import pyplot as plt
from control import *
```

```python
Y1, Y2 = [10, 10]
c=2

Q = np.array([[1, 0],
              [0, 0]])

U = np.array(1)

# W = np.array([[1, 0],
#               [0, 1]])

A = np.array([[ 0, 1],
              [-1, 0]])

B = np.array([[-1],
              [0]])

# R = np.array([-Y2, Y1])

K, S, E = lqr(A, B, Q, U)
K
```

```python
def fun(t, X, A, B, u):
    dx = A@X + B*u
    return dx[0]

B.T[0]*K
```

```python
x0 = np.array([Y1, Y2+c])

sol = scipy.integrate.solve_ivp(fun, [0,10],x0, t_eval=np.linspace(0,10, 1000), args=(A, B.T[0], K))
sol
```

```python
plt.figure()
plt.plot(sol.t, sol.y[0])
plt.show()
```

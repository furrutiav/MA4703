```python
import numpy as np
import control
```

Asignamos los valores de los parametros

```python
i0 = 1
L1 = 3.5
L2 = 2
Rc = 1
Rw = 3
alpha = 0.9
a = 50
```

Calculamos las matrices $A$ y $B$ del sistema

```python
const = 1/((1-alpha*alpha)*L1*L2)
K = alpha * np.sqrt(L1*L2) 
A = const * np.array([
    [-L2*Rc, alpha*np.sqrt(L1*L2)*Rw],
    [alpha*np.sqrt(L1*L2)*Rc, -L1*Rw]
])
B = const * np.array([[L2], [-K]])
```

Calculamos la matriz de Kalman $K$ del sistema $(A, B)$ y verificamos que es de rango completo

```python
kalman = control.ctrb(A, B)
rank_kalman = np.linalg.matrix_rank(kalman)
print("Matriz de Kalman del sistema (A, B):")
print(kalman)
print("Rango de la matriz:", rank_kalman)
```

Concluimos que el sistema $(A, B)$ es controlable

```python
from scipy.integrate import solve_ivp
from matplotlib import pyplot as plt
from matplotlib import rc
import seaborn as sns

plt.style.use(['seaborn', "seaborn-dark"][0])
plt.rc('axes', titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rcParams.update({'font.size': 16})
plt.rcParams['axes.titlesize'] = 16
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams.update({'lines.markeredgewidth': 1})
plt.rcParams.update({'errorbar.capsize': 2})
```

Calculamos la norma de la derivada para las trayectorias obtenidas al resolver la EDO con un control constante

```python
def norm_derivative_solve_u_cte(cte=0.5):
    u_cte = lambda t: cte
    sol_u = solve_ivp(lambda t, x: x @ A + B.T * u_cte(t), [0, 50], [-i0, 0], max_step=0.01)
    dx1 = (sol_u.y[0,:-1] - sol_u.y[0,1:])/(sol_u.t[:-1]-sol_u.t[1:])
    dx2 = (sol_u.y[1,:-1] - sol_u.y[1,1:])/(sol_u.t[:-1]-sol_u.t[1:])
    return sol_u.t[:-1], np.sqrt(dx1**2 + dx2**2)
```

Graficamos la norma de la derivada en funcion del tiempo, para controles constantes $[0, .5, 1, 2, 5]$

```python
fig, ax = plt.subplots(figsize=(10, 5))

for cte in [0, .5, 1, 2, 5]:
    T, N = norm_derivative_solve_u_cte(cte=cte)
    ax.plot(T, N, label=r"u = "+str(cte))

ax.set_ylabel(r'$||\dot{x}(t)||$', fontsize=12)
ax.set_xlabel('tiempo ($t$)', fontsize=12)
ax.set_title("Norma de la derivada en función del tiempo")
ax.legend();
plt.savefig("e3_1.pdf", bbox_inches='tight', )
```

## Ejercicio 4

El problema se puede escribir como

$$
min_{T, u} T
$$

sujeto a 

$$
\begin{align*}
i' &= Ai + Bu\\
i(T; i(0), u) &= (i_0, 0)^T\\
i(0) &= (-i_0, 0)^T\\
u &\in [-a, a]
\end{align*}
$$

## Ejercicio 5

```python
def euler(A, B, u, I0, tf, N):
    I = I0
    dt = tf/N
    solution = [I]
    for it in range(N):
        I = I + dt * (A @ I + u[it] * B)
        solution.append(I)
    return solution
```

```python
from scipy.optimize import minimize, NonlinearConstraint, LinearConstraint
```

```python
def solve_min_time(x0, N=100, eps = 1e-6, method="COBYLA"):
    def g(x):
        global A, B, i0, N
        tf, u = x[0], x[1:]
        I0 = np.array([[-i0], [0]])
        sol = euler(A, B, u, I0, tf, N)
        I = sol[-1]
        return I[0,0], I[1, 0]
    
    xf = np.array([i0, 0])
    sol_N = minimize(
        lambda x: x[0],
        x0 = x0,
        method="COBYLA",
        constraints=[
            NonlinearConstraint(g, xf-eps, xf+eps), 
            LinearConstraint(np.eye(N+1), 
                             lb=np.array([0]+[-a]*N), 
                             ub=np.array([np.inf]+[a]*N)
                            )],
        options = {"maxiter": 20000}
    )
    If = g(sol_N.x)
    return sol_N, If
```

```python
%%time
N=50
x0 = np.array([0.01]+[0]*N)
sol_N, If = solve_min_time(x0, N=N, eps=1e-6, method="COBYLA")
sol_N, If
```

La solución obtenida satisface que llega cerca de la posicion final deseada, sin embargo el control asociado no es extremal. Esto puede deberse a la mala elección del punto inicial (compatibilidad de tf0 y u0) y los parametros elegidos (N y eps)

## Ejercicio 7

```python
from scipy.optimize import root
```

```python
def F(x):
    tf, h = x[0], x[1]
    p = solve_ivp(lambda t, p: -A.T @ p, t_span=[0, tf], y0=[-1, h], max_step=0.001, dense_output=True)  
    u = lambda t: a * np.sign(p.sol(t).T @ B)[0]
    x = solve_ivp(lambda t, x: x @ A + u(t)*B.T, t_span=[0, tf], y0=[-i0, 0], max_step=0.001)
    return x.y[0,-1]-i0, x.y[1,-1]
```

```python
%%time
R = root(F, x0=(10,1), method="krylov")
R
```
